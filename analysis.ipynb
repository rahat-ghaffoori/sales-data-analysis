# Sales Data Analysis Project

## Objective
Analyse retail sales data to identify trends, customer behaviour, and key revenue drivers to support data-driven business decisions.

## 1. Import & Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("train.csv")
df.head()
```

## 2. Data Exploration

```python
df.info()
df.describe()
df.isnull().sum()
```

## 3. Data Cleaning

```python
# Convert date column
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df = df.dropna()
```

## 4. Feature Engineering

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

## 5. Data Visualisation

Sales Over Time

```python
import matplotlib.pyplot as plt

df.groupby('Year')['Sales'].sum().plot(kind='line')
plt.title("Total Sales Over Time")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

**Insight:** Sales show a steady upward trend over time, indicating consistent business growth and increasing demand.

Monthly Sales Trend

```python
df.groupby('Month')['Sales'].sum().plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

**Insight:** Sales vary across months, suggesting seasonal patterns in customer demand.

Sale by Category

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

**Insight:** Certain categories dominate total sales, suggesting they are the primary drivers of revenue.

Sales by Sub-Category

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values().plot(kind='barh')
plt.title("Sales by Sub-Category")
plt.xlabel("Total Sales")
plt.ylabel("Sub-Category")
plt.show()
```

**Insight:** Some sub-categories underperform, indicating opportunities for improvement.

Sales by Region

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

**Insight:** Sales are concentrated in specific regions, reflecting differences in market demand.

Sales by Customer Segment

```python
df.groupby('Segment')['Sales'].sum().plot(kind='bar')

plt.title("Sales by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales")
plt.show()
```

**Insight:** One customer segment contributes the majority of revenue, highlighting key target customers.

Top 10 Products by Sales

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='barh')

plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.show()
```

**Insight:** A small number of products generate a large portion of total sales.

## Advanced Insight

Sales are highly concentrated in specific regions, customer segments, and products, suggesting that the business relies on a limited number of key revenue drivers rather than a fully diversified sales strategy.

## Business Recommendations

- Focus marketing efforts on high-performing regions such as the West to maximise revenue
- Prioritise the Consumer segment for targeted campaigns and retention strategies
- Promote top-selling products to sustain revenue growth
- Improve performance of underperforming sub-categories through pricing or marketing adjustments
- Leverage seasonal trends to optimise inventory and sales planning

## Conclusion

This analysis provides insights into sales trends, customer behaviour, and product performance. The findings can support strategic decision-making and help improve overall business performance through targeted actions.
