# Sales Data Analysis

## 1. Data Loading

```python
import pandas as pd

df = pd.read_csv("train.csv")

df.head()
```

## 2. Data Exploration

```python
df.info()
df.describe()
df.isnull().sum()
```

## 3. Data Cleaning

```python
# Convert date column
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

# Check duplicates
df.duplicated().sum()

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df = df.dropna()
```

## 4. Feature Engineering

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
```

## 5. Data Visualisation

```python
import matplotlib.pyplot as plt

df.groupby('Year')['Sales'].sum().plot(kind='line')
plt.title("Total Sales Over Time")
plt.show()
```

**Insight:** Sales show an increasing trend over time, indicating business growth.

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

**Insight:** One category dominates sales, suggesting it is the main revenue driver.

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

**Insight:** Some regions are significantly more profitable than others.
